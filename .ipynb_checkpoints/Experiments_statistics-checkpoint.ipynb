```python
import pickle
from deap import tools
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from scipy import stats
```

```python
enemy_ids = [1, 2, 4]
runs = 10
```

## Line plots

```python
def read_logbook(enemy_id, runs):    
    logbooks_b = []
    logbooks_r = []

    for run in range(runs):

        with open(f'logBook/biased/enemy_{enemy_id}/logBook_run_{run}.pkl', 'rb') as f:
            logBook = pickle.load(f)

        logBook = pd.DataFrame.from_dict(logBook)
        logBook['run'] = f'{run}'

        logbooks_b.extend([logBook])

        with open(f'logBook/random/enemy_{enemy_id}/logBook_run_{run}.pkl', 'rb') as f:
            logBook = pickle.load(f)

        logBook = pd.DataFrame.from_dict(logBook)
        logBook['run'] = f'{run}'

        logbooks_r.extend([logBook])

    logbook_biased = pd.concat(logbooks_b)
    logbook_random = pd.concat(logbooks_r)

    return logbook_biased, logbook_random
```

```python
def training_plot(enemy_id, logbook_random, logbook_biased, ax):

    # set fixed y-axis
    ax.set_ylim([0, 100])

    lineplots = sns.lineplot(data=logbook_random, x='gen', y='max', errorbar='sd', \
        color='b', label='random_max', ax=ax)
    lineplots = sns.lineplot(data=logbook_random, x='gen', y='mean', errorbar='sd', \
        color='b', label='random_mean', dashes=[2,2], err_style='band', ax=ax)
    lineplots = sns.lineplot(data=logbook_biased, x='gen', y='max', errorbar='sd', \
        color='r', label='biased_max', ax=ax)
    lineplots = sns.lineplot(data=logbook_biased, x='gen', y='mean', errorbar='sd', \
        color='r', label='biased_mean', dashes=[2,2], err_style='band', ax=ax)

    lineplots.tick_params(axis='x', which='major', bottom=True, top=False, labelbottom=False)

    lineplots.set_xlabel('', fontsize=20)
    lineplots.set_ylabel('fitness value', fontsize=20)
    lineplots.axes.set_title(f'Mean/Max for enemy {enemy_id}', fontsize=20)
    lineplots.tick_params(labelsize=15)
    ax.legend(fontsize=18, loc='lower right').remove()
```

```python
plt.rcParams['figure.figsize'] = (6, 20)
fig_lineplot, ax_lineplot = plt.subplots(3, 1)

for i, enemy_id in enumerate(enemy_ids):
    logbook_biased, logbook_random = read_logbook(enemy_id, runs)
    training_plot(enemy_id, logbook_random, logbook_biased, ax=ax_lineplot[i])
ax_lineplot[1].legend(fontsize=18, loc='lower right')

ax_lineplot[2].set_xlabel('Generations', fontsize=20)
ax_lineplot[2].tick_params(axis='x', which='major', bottom=True, top=True, \
    labelbottom=True)

plt.show()
```

```python
# save lineplots
fig_line = fig_lineplot.get_figure()
fig_line.savefig('lineplot_figures/training_statistics.pdf', bbox_inches='tight')
```

## Box plots

```python
def read_test_results(enemy_id):
    test_biased = np.loadtxt(f'test_results/biased/enemy_{enemy_id}.txt')
    df1 = pd.DataFrame(test_biased, columns=['biased']).assign(Trial=1)
    test_random = np.loadtxt(f'test_results/random/enemy_{enemy_id}.txt')
    df2 = pd.DataFrame(test_random, columns=['random']).assign(Trial=2)

    df_test = pd.concat([df1, df2])
    df_test = pd.melt(df_test, id_vars=['Trial'], var_name=['Number'])  

    return df_test
```

```python
def test_plot(enemy_id, df_test, ax):

    # set fixed y-axis
    ax.set_ylim([-20, 100]) 

    boxplots = sns.boxplot(x='Trial', y='value', hue='Number', data=df_test, dodge=False, ax=ax)

    boxplots.set_xlabel('', fontsize=20)
    boxplots.set_ylabel('individual gain', fontsize=20)
    boxplots.axes.set_title(f'Individual gains for enemy {enemy_id}', fontsize=20)
    # boxplots.set_xticklabels(['Biased', 'Random'])
    boxplots.tick_params(axis='x', which='major', bottom=True, top=False, labelbottom=False)
    ax.legend().remove()
```

```python
plt.rcParams['figure.figsize'] = (6, 20)
fig_boxplot, ax_boxplot = plt.subplots(3, 1)

for i, enemy_id in enumerate(enemy_ids):
    df_test = read_test_results(enemy_id)
    test_plot(enemy_id, df_test, ax=ax_boxplot[i])

ax_boxplot[2].set_xticklabels(['Biased', 'Random'])
ax_boxplot[2].tick_params(axis='x', which='major', bottom=True, top=True, \
    labelbottom=True, labelsize=20)

plt.show()
```

```python
# save boxplots
fig_box = fig_boxplot.get_figure()
fig_box.savefig(f'boxplot_figures/test_boxplot.pdf', bbox_inches='tight')
```

## Statistical significance test (Welch's t-test)

```python
print("The hypothesis that the mean of the two algorithms is equal is rejected if p < 0.05.")

for enemy_id in enemy_ids:
    test_biased = np.loadtxt(f'test_results/biased/enemy_{enemy_id}.txt')
    test_random = np.loadtxt(f'test_results/random/enemy_{enemy_id}.txt')
    t, p = stats.ttest_ind(test_biased, test_random, equal_var=False)
    
    print(f"\nEnemy {enemy_id}:\nt = {t}\np = {p}, so the difference is{' NOT ' if p > 0.05 else ' '}significant!\n")
```
